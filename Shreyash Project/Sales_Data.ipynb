```python
# Detect numeric columns for Y-axis
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

# Detect first non-numeric (text/date) column for X-axis
non_numeric_cols = df.select_dtypes(exclude=np.number).columns.tolist()

# Set default X and Y axis
x_col = non_numeric_cols[0] if non_numeric_cols else df.columns[0]
y_cols = numeric_cols if numeric_cols else df.columns[1:]

print(f"\nX-axis: {x_col}")
print(f"Y-axis: {y_cols}")
```

```python
# 1. Column Chart (non-stacked)
df.plot(x=x_col, y=y_cols, kind='bar', title='Column Chart (Non-Stacked)', figsize=(10,6))
plt.show()




```

```python

# 2. Column Chart (stacked)
df.plot(x=x_col, y=y_cols, kind='bar', stacked=True, title='Column Chart (Stacked)', figsize=(10,6))
plt.show()


```

```python
# 3. Bar Chart (horizontal)
df.plot(x=x_col, y=y_cols, kind='barh', title='Bar Chart', figsize=(10,6))
plt.show()
```

```python

# 4. Area Chart (non-stacked)
df.plot(x=x_col, y=y_cols, kind='area', title='Area Chart (Non-Stacked)', figsize=(10,6))
plt.show()
```

```python
# 5. Area Chart (stacked)
df.plot(x=x_col, y=y_cols, kind='area', stacked=True, title='Area Chart (Stacked)', figsize=(10,6))
plt.show()
```

```python
# 6. Line Chart
df.plot(x=x_col, y=y_cols, kind='line', title='Line Chart', figsize=(10,6))
plt.show()
```

```python
# 7. Scatter Chart (XY)
if len(y_cols) >= 2:
    plt.scatter(df[y_cols[0]], df[y_cols[1]])
    plt.title('XY (Scatter) Chart')
    plt.xlabel(y_cols[0])
    plt.ylabel(y_cols[1])
    plt.show()
```

```python
# 8. Bubble Chart (3D version using size)
if len(y_cols) >= 3:
    plt.scatter(df[y_cols[0]], df[y_cols[1]], s=df[y_cols[2]]*10, alpha=0.5)
    plt.title('Bubble Chart')
    plt.xlabel(y_cols[0])
    plt.ylabel(y_cols[1])
    plt.show()

```

```python
# 9. Donut Chart (Pie chart with hole)
if len(y_cols) >= 1:
    data = df[y_cols[0]].head(5)
    labels = df[x_col].head(5)
    plt.pie(data, labels=labels, wedgeprops={'width':0.4})
    plt.title('Donut Chart')
    plt.show()
```

```python
# 10. Column + Line Combo Chart
fig, ax1 = plt.subplots(figsize=(10,6))
df.plot(x=x_col, y=y_cols[0], kind='bar', ax=ax1, color='skyblue', label=y_cols[0])
ax2 = ax1.twinx()
df.plot(x=x_col, y=y_cols[1] if len(y_cols) > 1 else y_cols[0], kind='line', ax=ax2, color='red', label='Line')
plt.title('Column and Line Combo Chart')
plt.show()
```

```python
# 11. Stock Chart (using line chart simulation)
df.plot(x=x_col, y=y_cols, kind='line', title='Stock Chart (Line Representation)', figsize=(10,6))
plt.show()
```
